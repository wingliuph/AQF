```python
import sys
import pandas as pd
#import tushare as ts
import os
import glob
import copy
import datetime
import numpy as np
from struct import *
import pymysql
```

```python
def jiexiTDX_daily ( jiaoyisuo= 'SH'):
    if(jiaoyisuo=='SH'):    
        j_add='D:/Program Files (x86)/tdx/vipdoc/sh/lday/'
        sql="insert into sh_daily (date_id,open1,high,low,close,amount,vol,code_id) values(%s,%s,%s,%s,%s,%s,%s,%s)"

    elif(jiaoyisuo=='SZ'):
        j_add='D:/Program Files (x86)/tdx/vipdoc/sz/lday/'
        sql="insert into sz_daily (date_id,open1,high,low,close,amount,vol,code_id) values(%s,%s,%s,%s,%s,%s,%s,%s)"

    else:
        print("等我编辑其他交易所数据获取方式。")

    os.chdir(j_add)
    sh_list = glob.glob('*.day')
    #address = os.getcwd()#返回当前目录
    for code in sh_list:
        conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='123', db='tdx')
        cursor = conn.cursor()
        dayfile = open(code, 'rb')
        buffer = dayfile.read()
        dayfile.close()
        code_id = code[2:8]+"."+jiaoyisuo
        #读取的数据长度
        buflengh = len(buffer) #32的整数倍
        #除32获利文件中存储了多少天的信息
        datenum = int(buflengh / 32)
        #解析每一天的数据的内容,每一天列表的起始位置和终止位置
        list_begin = 0
        list_end = 32
        #创建pandas股票数据表
        #将个股数据转换为pandas的dataFrame格式数据
        dl =[]
        for K in range(datenum):
            #解析32个字节一天数据
            a = unpack('IIIIIfII', buffer[list_begin:list_end])
            date_id  = pd.to_datetime(str(a[0]),format='%Y-%m-%d').date()
            open1 = a[1] / 100.0
            high  = a[2] / 100.0
            low   = a[3] / 100.0
            close = a[4] / 100.0
            amount = a[5] / 10.0
            vol = a[6]
            unused = a[7]
            #dl.append([date_id,open1,high,low,close,amount,vol,code_id])
            dl=[date_id,open1,high,low,close,amount,vol,code_id]
            #加32指向下一天的数据列表
            list_begin = list_begin + 32
            list_end = list_end + 32
            cursor.execute(sql,dl)# 每日的数据添加一次，不再汇总后添加
            conn.commit() 
        cursor.close()     # 关闭游标
        conn.close()      # 关闭连接
        
        print (code_id + ' ，has been saved in MYSQL.TDX的 '+jiaoyisuo+' 交易所中!' )
```

```python
if __name__ == "__main__":
    jiexiTDX_daily(jiaoyisuo='SZ')
    jiexiTDX_daily(jiaoyisuo='SH')
```
